# **Crunchbase Companies in the SAAS Industry**

***Approach to Scraping:***

> The scraping process involves leveraging BeautifulSoup and the requests library within a Colab environment to extract company data from Crunchbase within the SAAS industry. The script initiates by fetching the base URL and parsing HTML content to retrieve links directing to individual company profiles. Iterating through these links, the script gathers extensive company details, encompassing name, description, location, employee count, funding rounds, IPO status, company URL, Crunchbase rank, hiring status, contacts count, and stock symbol.





***Challenges and Solutions:***

1. **Dynamic Content Handling:** Crunchbase's dynamic content requires managing pauses between requests (`time.sleep()`). This strategy emulates human behavior, ensuring data retrieval without triggering bot detection mechanisms.

2. **Varied Data Extraction Methods:** Diverse strategies, including the use of various BeautifulSoup methods (`find()`, `find_all()`, `select()`), with different CSS selectors, enable precise extraction of data from diverse HTML elements and attributes on the webpage.

3. ***Overcoming Anti-Bot Measures:*** Initial consideration of Selenium for dynamic element handling encountered PrimeterX captcha, prompting a shift back to BeautifulSoup. The emphasis lies in cautious parsing and behavior mimicking to evade detection, considering the detectability of browser automation tools. Even after implementing random wait times, prolonged scraping sessions or larger datasets can still trigger defensive measures.

***Integration of Rotating Proxies:***

Despite implementing random wait times between requests, we encountered defensive measures after extended scraping sessions or with larger datasets. To overcome this, we acknowledge the effectiveness of rotating proxies. These proxies offer dynamic IP address changes, significantly reducing the risk of detection by diversifying request sources.

> While the challenges of acquiring free IP addresses or the costs of paid services were considered, their application, especially when dealing with larger datasets, proves invaluable. Rotating proxies not only optimize performance and minimize wait times but also ensure a smoother, uninterrupted extraction process. By systematically switching IPs, they enhance efficiency, allowing for continuous data extraction without interruptions due to defensive measures.

***Scaling the Project:***

Efficient Iteration and Handling Limits: Exploring parallel processing techniques using libraries like `concurrent.futures` aims to optimize the script for larger datasets. This approach can enhance extraction efficiency while ensuring compliance with Crunchbase's terms of service and mitigating risks associated with rate limits.

***Additional Techniques and Tools:***

- **Enhancing Data Extraction:** Continued exploration of parsing techniques and CSS selectors within BeautifulSoup remains a potential avenue for refining the extraction process, improving data accuracy, and enhancing extraction efficiency within the Colab environment.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import time
import random

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36",
}

url = 'https://www.crunchbase.com/discover/organization.companies/c57ddc9ae990522c47aaf98455dd8c26'
soup = BeautifulSoup(requests.get(url, headers=headers).text, 'html.parser')
identifier_links = soup.find_all('a', {'href': lambda href: href and href.startswith('/organization')})

data = []
company_data = []
for link in identifier_links:
    time.sleep(random.uniform(3, 10))
    company_url = 'https://www.crunchbase.com' + link.get('href')
    company_soup = BeautifulSoup(requests.get(company_url, headers=headers).text, 'html.parser')
    time.sleep(random.uniform(3, 5))

   # Find the h1 tag with class 'profile-name'
    company_name = company_soup.find('h1', {'class': 'profile-name'}).get_text(strip=True) if company_soup.find('h1', {'class': 'profile-name'}) else None
    # DESCRIPTOIN OF THE COMPANY
    meta_tag = company_soup.find('meta', {'property': 'og:description'})
    og_description = meta_tag.get('content') if meta_tag else None
    #LOCATION OF THE COMPANY
    location_span = company_soup.find('span', {'class': 'component--field-formatter'})
    location = location_span.get_text(separator=',', strip=True) if location_span else None
    #NUMBER OF EMPLYEES
    employees_link = company_soup.find('a', {'class': 'component--field-formatter'})
    number_of_employees = employees_link.get_text(strip=True) if employees_link else None
    #IPO STATUS
    pattern = re.compile(r'/search/funding_rounds/field/organizations/last_funding_type/')
    matching_a_tags = company_soup.find('a', href=pattern)
    data_funding_rounds = [a_tag.get_text(strip=True) for a_tag in matching_a_tags] if matching_a_tags else None
    time.sleep(random.uniform(3, 5))
    ipo_status = company_soup.find('span', {'class': 'component--field-formatter field-type-enum ng-star-inserted'})
    ipo_status_text = ipo_status.get_text(strip=True) if ipo_status else None
    #company URL
    company_url_tag = company_soup.find('a', {'class': 'component--field-formatter accent ng-star-inserted'})
    company_url_text = company_url_tag.get_text(strip=True) if company_url_tag else None
    #CB RANK
    field_formatters = company_soup.find_all('field-formatter')
    cb_rank = None
    for field_formatter in field_formatters:
        a_tag = field_formatter.find('a', {'class': 'component--field-formatter field-type-integer accent highlight-color-contrast-light ng-star-inserted'})
        if a_tag:
            cb_rank = a_tag.get_text(strip=True)
            break
        time.sleep(random.uniform(3, 5))
    #HIRING OR NOT
    link_formatters = company_soup.find_all('link-formatter')
    actively_hiring = False
    for link_formatter in link_formatters:
        if 'Actively Hiring - View All Jobs' in link_formatter.get_text(strip=True):
            actively_hiring = True
            break
    #CONTACTS COUNT
    label_elements = company_soup.find_all('span', {'class': 'wrappable-label-with-info ng-star-inserted'})
    contact_count = None
    for label_element in label_elements:
        label_text = label_element.get_text(strip=True)
        if 'Contacts' in label_text:
            contact_span = label_element.find_next('span', {'class': 'component--field-formatter field-type-integer ng-star-inserted'})
            contact_count = contact_span.get_text(strip=True) if contact_span else None
            break

    #STOCK SYMBOL
    stock_symbol = None
    for label_element in label_elements:
        label_text = label_element.get_text(strip=True)
        if 'Stock Symbol' in label_text:
            stock_symbol_element = label_element.find_next('link-formatter')
            if stock_symbol_element:
                stock_symbol = stock_symbol_element.get_text(strip=True)
                break
    if stock_symbol is None:
        stock_symbol = 'No Stock Symbol'


    #APPEND DATA
    company_data.append({
        'Organization': company_name,
        'og_description': og_description,
        'location': location,
        'number_of_employees': number_of_employees,
        'data_funding_rounds': data_funding_rounds,
        'ipo_status': ipo_status_text,
        'company_url': company_url_text,
        'cb_rank': cb_rank,
        'actively_hiring': 'Actively Hiring' if actively_hiring else 'Not hiring',
        'Contacts Num': contact_count,
    })

    # Introduce a delay to mimic human behavior
    time.sleep(random.uniform(3, 5))  # Adjust the delay time as needed

# Create DataFrame
df_f = pd.DataFrame(company_data)
```

# TESTS USING PROXIES -- FOR DECLARATION PURPOSE

>The script utilized a rotating proxy method to access the website to prevent detection by defensive systems. However, it's important to note the limitations of using free, random proxies, which resulted in intermittent failure. The script attempted to cycle through multiple proxies, but due to the unpredictability and quality of these free proxies, not all requests were successful. This approach is a trade-off between anonymity and reliability, and in this case, it resulted in inconsistent access to the website.



```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from itertools import cycle


headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36",
}
url = 'https://www.crunchbase.com/discover/organization.companies/c57ddc9ae990522c47aaf98455dd8c26'


rotating_proxies = [
    '196.203.83.249:9090',
    '102.219.178.24:3128',
    '102.219.178.143:3128',
    '41.231.54.99:80',
    '197.14.14.238:80',
    '196.203.30.85:8080',
    '197.13.10.14:3128',
    '102.219.178.236:3128',
    '102.219.179.58:3128',
    '102.219.179.92:3128',
    '196.179.204.9:8080',
    '102.219.177.38:3128',
    '41.229.253.214:8080',
    '41.231.54.37:8888',
    '41.230.30.147:3128',
    '197.14.14.234:80'
]
rotating_proxy_pool = cycle(rotating_proxies)
data = []

try:
    working_proxy = None
    for rotating_proxy in rotating_proxy_pool:
        proxy_dict = {"http": rotating_proxy, "https": rotating_proxy}
        try:
            response = requests.get(url, headers=headers, proxies=proxy_dict, timeout=60)
            if response.status_code == 200:
                working_proxy = rotating_proxy_pool
                print(f"Proxy {rotating_proxy} is working.")
                break
            print(f"Proxy {rotating_proxy} failed with status code: {response.status_code}")
        except Exception as e:
            print(f"Error with proxy {rotating_proxy}: {e}")
    if working_proxy:
        print(f"Using rotating proxy: {working_proxy} for the initial request")
    else:
        print("No working rotating proxies found.")
except Exception as e:
    print(f"Error: {e}")

```
